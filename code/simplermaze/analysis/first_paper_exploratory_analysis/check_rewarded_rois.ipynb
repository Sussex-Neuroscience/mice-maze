```python
results = []
overall_hits = 0
all_mouse_hits = []

for path in trials_csv_paths:
    dir_name = os.path.dirname(path)[-3:]
    df = pd.read_csv(path)

    # 1. Filter only the hits (Vectorized)
    hits_df = df[df["hit"] == 1].copy()
    counter_hits = len(hits_df)
    
    if counter_hits == 0:
        continue # Avoid division by zero if a session has no hits

    # 2. Extract ROI counts efficiently
    # This maps 'rew_location' to counts for A, B, C, D
    counts = hits_df["rew_location"].value_counts()
    
    # 3. Calculate probabilities using a dictionary comprehension
    rois = ['A', 'B', 'C', 'D']
    probs = {roi: counts.get(roi, 0) / counter_hits for roi in rois}
    
    # 4. Store session data
    results.append({
        'session': dir_name,
        'total_hits': counter_hits,
        **probs  # Unpacks the A, B, C, D probabilities
    })

    # Prepare data for the master mouse_hits DataFrame
    hits_df['session'] = dir_name
    all_mouse_hits.append(hits_df[['session', 'rew_location']])
    
    overall_hits += counter_hits

    #just get rois visited in all the trials

# Create a clean summary DataFrame
summary_df = pd.DataFrame(results)
mouse_hits_df = pd.concat(all_mouse_hits, ignore_index=True)

summary_df
```
