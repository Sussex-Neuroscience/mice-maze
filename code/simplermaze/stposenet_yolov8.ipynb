# Pose estimation 

## we are going to use STPoseNet (not DLC)

Spatiotemporal model built on the YOLOv8 architecture. Unlike models that analyze each frame in isolation, STPoseNet uses information from previous frames to improve the detection in the current frame.

The model's core is a time-domain tracking strategy with two main components:

- Tracking-Cropping Module: The model uses the key-point detection results from the previous frame to identify potential target locations in the next frame. This helps the model stay "locked on" to the mouse.

- Kalman Filter Integration: A Kalman filter is used to analyze the movement of key-points over time. This allows the model to reliably estimate the position of key-points even if they are temporarily missing (e.g., occluded).

Transfer learning can be used, allowing the model to achieve accurate detection even with a small training dataset.
