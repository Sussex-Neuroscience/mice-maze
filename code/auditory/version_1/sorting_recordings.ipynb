```python
import os
import shutil

drive = r"D:\Auditory data"
tem = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w1_d1"
i_one = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w1_d2"
i_two = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w1_d3"
pure_i = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w1_d4"
seq = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w2_sequences"
voc = r"C:\Users\shahd\Box\Awake Project\Maze data\Auditory experiments\8_arms_w_voc\w2_vocalisations"

for dirpath, dirnames, filenames in os.walk(drive):
    # Skip directories that we have already sorted
    if "dopaminergic" in dirpath.lower() or "maze" in dirpath.lower():
        continue

    folder_name = os.path.basename(dirpath)

    for file in filenames: 
        target_destination = None
        
        # Check which category this folder belongs to based on the file name
        if "temporal_envelope_modulation" in file:
            target_destination = os.path.join(tem, folder_name)
        elif "intervals_pt1" in file:
            target_destination = os.path.join(i_one, folder_name)
        elif "intervals_pt2" in file:
            target_destination = os.path.join(i_two, folder_name)
        elif "pure_intervals" in file:
            target_destination = os.path.join(pure_i, folder_name)
        elif "vocalisations" in file:
            target_destination = os.path.join(voc, folder_name)
        elif "sequences" in file:
            target_destination = os.path.join(seq, folder_name)

        # copy if we found a matching keyword
        if target_destination:
            try:
                shutil.copytree(dirpath, target_destination, dirs_exist_ok=True)
                print(f"Successfully copied: {folder_name} to {target_destination}")
                
                break 
            except Exception as e:
                print(f"Error copying {dirpath}: {e}")
                break # Stop trying this folder if an error occurs
```
