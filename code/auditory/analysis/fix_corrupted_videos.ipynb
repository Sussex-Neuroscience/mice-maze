```python
# Jupyter-friendly MP4 repair for a *list* of file paths (videostofix).
# Run this single cell. If you already have `videostofix = [ ... ]` defined,
# it will process those files. Otherwise you'll get a reminder to define it.
#
# Requirements: ffmpeg + ffprobe on PATH. OpenCV optional (pip install opencv-python-headless).

import os
import csv
import json
import shutil
import tempfile
import subprocess
from pathlib import Path
from dataclasses import dataclass, asdict
from typing import Optional, Tuple, List

# Optional: nice DataFrame display if pandas is available
try:
    import pandas as pd
    from caas_jupyter_tools import display_dataframe_to_user
    PANDAS_OK = True
except Exception:
    PANDAS_OK = False

# ---------- Settings (you can edit these defaults) ----------
FPS_FALLBACK = 30          # used when timestamps are broken / fps=0
KEEP_INTERMEDIATES = False # keep temp files for debugging
AVI_FALLBACK = False       # also try producing an MJPG .avi as last resort
REPORT_PATH = Path("repair_report_from_list.csv")  # aggregated CSV report path
# ------------------------------------------------------------


# ---------- Helpers ----------
def which(cmd: str) -> Optional[str]:
    from shutil import which as _which
    return _which(cmd)

def run(cmd: List[str], timeout: Optional[int] = None) -> Tuple[int, str, str]:
    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    try:
        out, err = p.communicate(timeout=timeout)
    except subprocess.TimeoutExpired:
        p.kill()
        out, err = p.communicate()
    return p.returncode, out, err

def ffprobe_json(input_path: Path) -> Tuple[Optional[dict], str]:
    if not which("ffprobe"):
        return None, "ffprobe not found on PATH"
    cmd = [
        "ffprobe", "-v", "error", "-print_format", "json",
        "-show_format", "-show_streams", str(input_path)
    ]
    code, out, err = run(cmd)
    if code != 0:
        return None, (err.strip() or f"ffprobe exit code {code}")
    try:
        info = json.loads(out)
        return info, ""
    except Exception as e:
        return None, f"ffprobe json parse error: {e}"

def cv2_validate(input_path: Path, test_frames: int = 30) -> Tuple[bool, str, int, float]:
    """
    Try to open and read a few frames with OpenCV.
    Returns (ok, reason, frames_read, fps_reported).
    """
    try:
        import cv2
    except Exception:
        return False, "OpenCV not installed", 0, 0.0

    cap = cv2.VideoCapture(str(input_path))
    if not cap.isOpened():
        return False, "cv2.VideoCapture failed to open", 0, 0.0
    fps = cap.get(cv2.CAP_PROP_FPS) or 0.0
    frames_ok = 0
    reason = "ok"
    for i in range(test_frames):
        ok, frame = cap.read()
        if not ok or frame is None:
            reason = f"read stopped at frame {i}"
            break
        frames_ok += 1
    cap.release()
    return frames_ok > 0, reason, frames_ok, float(fps)

def try_remux(input_path: Path, out_path: Path, fps_fallback: int) -> Tuple[bool, str]:
    """
    Fast remux: copy streams, rebuild timestamps, and move moov atom to front.
    """
    if not which("ffmpeg"):
        return False, "ffmpeg not found on PATH"
    # First attempt: straight copy + faststart
    cmd = [
        "ffmpeg", "-v", "error",
        "-fflags", "+genpts",
        "-i", str(input_path),
        "-c", "copy",
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, "remux ok"
    # Second attempt: force nominal framerate (some builds ignore with -c copy, but worth trying)
    cmd = [
        "ffmpeg", "-v", "error",
        "-fflags", "+genpts",
        "-r", str(fps_fallback),
        "-i", str(input_path),
        "-c", "copy",
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, "remux ok (forced r)"
    return False, err.strip() or "remux failed"

def try_reencode(input_path: Path, out_path: Path, fps_fallback: int, preset: str = "veryfast") -> Tuple[bool, str]:
    """
    Re-encode video to H.264, ignore audio. This often rescues files with bad timestamps.
    """
    if not which("ffmpeg"):
        return False, "ffmpeg not found on PATH"
    cmd = [
        "ffmpeg", "-v", "error",
        "-fflags", "+genpts",
        "-i", str(input_path),
        "-an",
        "-c:v", "libx264",
        "-preset", preset,
        "-pix_fmt", "yuv420p",
        "-vsync", "2",
        "-r", str(fps_fallback),
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, "reencode ok"
    return False, err.strip() or "reencode failed"

def try_extract_rebuild(input_path: Path, out_path: Path, fps_fallback: int, tmp_dir: Path) -> Tuple[bool, str]:
    """
    Last-resort: extract frames (best-effort) then rebuild a video at fallback FPS.
    """
    if not which("ffmpeg"):
        return False, "ffmpeg not found on PATH"
    frames_dir = tmp_dir / "frames"
    frames_dir.mkdir(parents=True, exist_ok=True)
    # Extract frames (ignore decode errors)
    cmd_extract = [
        "ffmpeg", "-v", "error",
        "-err_detect", "ignore_err",
        "-i", str(input_path),
        "-vsync", "0",
        str(frames_dir / "%06d.png")
    ]
    code, out, err = run(cmd_extract)
    extracted = sorted(frames_dir.glob("*.png"))
    if len(extracted) == 0:
        return False, err.strip() or "no frames extracted"
    # Rebuild at fixed FPS
    cmd_build = [
        "ffmpeg", "-v", "error",
        "-framerate", str(fps_fallback),
        "-i", str(frames_dir / "%06d.png"),
        "-c:v", "libx264",
        "-pix_fmt", "yuv420p",
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd_build)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, f"rebuilt from {len(extracted)} frames"
    return False, err.strip() or "rebuild failed"

def try_avi_fallback(input_path: Path, out_path: Path, fps_fallback: int) -> Tuple[bool, str]:
    """
    Optional: produce an MJPG AVI as a compatibility fallback.
    """
    if not which("ffmpeg"):
        return False, "ffmpeg not found on PATH"
    cmd = [
        "ffmpeg", "-v", "error",
        "-fflags", "+genpts",
        "-i", str(input_path),
        "-an",
        "-c:v", "mjpeg",
        "-qscale:v", "3",
        "-r", str(fps_fallback),
        str(out_path)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, "avi fallback ok"
    return False, err.strip() or "avi fallback failed"


@dataclass
class FileReport:
    source: str
    ok_original: bool
    reason_original: str
    frames_read_original: int
    fps_reported_original: float
    probe_error: str
    action: str
    ok_fixed: bool
    reason_fixed: str
    output_file: str
    notes: str


def process_file_to_local_fixed(src: Path,
                                fps_fallback: int = 30,
                                keep_intermediates: bool = False,
                                try_avi: bool = False) -> FileReport:
    """
    Process a single file and write output to a 'fixed' subfolder next to the source.
    """
    if not src.exists():
        return FileReport(
            source=str(src),
            ok_original=False,
            reason_original="file not found",
            frames_read_original=0,
            fps_reported_original=0.0,
            probe_error="",
            action="skipped",
            ok_fixed=False,
            reason_fixed="input missing",
            output_file="",
            notes=""
        )

    out_dir = src.parent / "fixed"
    out_dir.mkdir(parents=True, exist_ok=True)

    ok_cv, reason_cv, frames_read, fps_rep = cv2_validate(src)
    info, probe_err = ffprobe_json(src)

    # If looks OK, still offer to remux to faststart
    if ok_cv and not probe_err:
        fixed = out_dir / (src.stem + ".faststart.mp4")
        ok_r, r_msg = try_remux(src, fixed, fps_fallback)
        action = "remux faststart" if ok_r else "none"
        return FileReport(
            source=str(src),
            ok_original=True,
            reason_original=reason_cv,
            frames_read_original=frames_read,
            fps_reported_original=fps_rep,
            probe_error="",
            action=action,
            ok_fixed=ok_r,
            reason_fixed=r_msg if ok_r else "original already ok",
            output_file=str(fixed if ok_r else src),
            notes=""
        )

    # Otherwise, staged repairs
    tmp = Path(tempfile.mkdtemp(prefix="mp4fix_"))
    try:
        # 1) Remux
        remux_path = tmp / (src.stem + ".remux.mp4")
        ok_r, r_msg = try_remux(src, remux_path, fps_fallback)
        if ok_r:
            ok_cv2, reason_cv2, frames2, fps2 = cv2_validate(remux_path)
            if ok_cv2:
                final = out_dir / (src.stem + ".fixed.remux.mp4")
                shutil.move(str(remux_path), str(final))
                if not keep_intermediates:
                    shutil.rmtree(tmp, ignore_errors=True)
                return FileReport(
                    source=str(src),
                    ok_original=False,
                    reason_original=reason_cv,
                    frames_read_original=frames_read,
                    fps_reported_original=fps_rep,
                    probe_error=probe_err,
                    action="remux",
                    ok_fixed=True,
                    reason_fixed="remux ok",
                    output_file=str(final),
                    notes=""
                )
        # 2) Re-encode
        reenc_path = tmp / (src.stem + ".reenc.mp4")
        ok_e, e_msg = try_reencode(src, reenc_path, fps_fallback)
        if ok_e:
            ok_cv3, reason_cv3, frames3, fps3 = cv2_validate(reenc_path)
            if ok_cv3:
                final = out_dir / (src.stem + ".fixed.h264.mp4")
                shutil.move(str(reenc_path), str(final))
                if not keep_intermediates:
                    shutil.rmtree(tmp, ignore_errors=True)
                return FileReport(
                    source=str(src),
                    ok_original=False,
                    reason_original=reason_cv,
                    frames_read_original=frames_read,
                    fps_reported_original=fps_rep,
                    probe_error=probe_err,
                    action="reencode",
                    ok_fixed=True,
                    reason_fixed="reencode ok",
                    output_file=str(final),
                    notes=""
                )
        # 3) Extract + rebuild
        rebuilt_path = tmp / (src.stem + ".rebuilt.mp4")
        ok_b, b_msg = try_extract_rebuild(src, rebuilt_path, fps_fallback, tmp)
        if ok_b:
            ok_cv4, reason_cv4, frames4, fps4 = cv2_validate(rebuilt_path)
            if ok_cv4:
                final = out_dir / (src.stem + ".rebuilt.mp4")
                shutil.move(str(rebuilt_path), str(final))
                if not keep_intermediates:
                    shutil.rmtree(tmp, ignore_errors=True)
                return FileReport(
                    source=str(src),
                    ok_original=False,
                    reason_original=reason_cv,
                    frames_read_original=frames_read,
                    fps_reported_original=fps_rep,
                    probe_error=probe_err,
                    action="extract+rebuild",
                    ok_fixed=True,
                    reason_fixed=b_msg,
                    output_file=str(final),
                    notes=""
                )
        # 4) Optional AVI fallback
        if try_avi:
            avi_path = tmp / (src.stem + ".fixed.avi")
            ok_a, a_msg = try_avi_fallback(src, avi_path, fps_fallback)
            if ok_a:
                final = out_dir / avi_path.name
                shutil.move(str(avi_path), str(final))
                if not keep_intermediates:
                    shutil.rmtree(tmp, ignore_errors=True)
                return FileReport(
                    source=str(src),
                    ok_original=False,
                    reason_original=reason_cv,
                    frames_read_original=frames_read,
                    fps_reported_original=fps_rep,
                    probe_error=probe_err,
                    action="avi-fallback",
                    ok_fixed=True,
                    reason_fixed="avi ok",
                    output_file=str(final),
                    notes=""
                )
        # Failed all
        return FileReport(
            source=str(src),
            ok_original=False,
            reason_original=reason_cv,
            frames_read_original=frames_read,
            fps_reported_original=fps_rep,
            probe_error=probe_err,
            action="failed-all",
            ok_fixed=False,
            reason_fixed="all repair attempts failed",
            output_file="",
            notes=r_msg or e_msg or b_msg or "n/a"
        )
    finally:
        if not keep_intermediates:
            shutil.rmtree(tmp, ignore_errors=True)


def notebook_repair_list(file_list: List[str],
                         fps_fallback: int = FPS_FALLBACK,
                         keep_intermediates: bool = KEEP_INTERMEDIATES,
                         avi_fallback: bool = AVI_FALLBACK,
                         report_path: Path = REPORT_PATH):
    """Repair the exact files in file_list. Outputs go to a 'fixed' subfolder next to each source file."""
    # Check tools
    if which("ffmpeg") is None or which("ffprobe") is None:
        print("WARNING: ffmpeg/ffprobe not found on PATH. Please install them before running repairs.")

    # Normalize and filter to existing files
    paths = [Path(p) for p in file_list]
    missing = [str(p) for p in paths if not p.exists()]
    if missing:
        print("Some paths do not exist and will be skipped:\n - " + "\n - ".join(missing))

    rows = []
    existing = [p for p in paths if p.exists() and p.suffix.lower() == ".mp4"]
    if not existing:
        print("No existing .mp4 files found in the provided list.")
        return None

    print(f"Processing {len(existing)} file(s)â€¦\n")
    for idx, src in enumerate(existing, 1):
        print(f"[{idx}/{len(existing)}] {src.name}")
        try:
            rep = process_file_to_local_fixed(
                src,
                fps_fallback=fps_fallback,
                keep_intermediates=keep_intermediates,
                try_avi=avi_fallback
            )
            rows.append(asdict(rep))
            status = "OK" if rep.ok_fixed or rep.ok_original else "FAIL"
            out_name = Path(rep.output_file).name if rep.output_file else "-"
            print(f"    -> {rep.action:14s} {status:4s}  {out_name}")
        except Exception as e:
            rows.append(asdict(FileReport(
                source=str(src),
                ok_original=False,
                reason_original=f"exception: {e}",
                frames_read_original=0,
                fps_reported_original=0.0,
                probe_error="",
                action="exception",
                ok_fixed=False,
                reason_fixed="exception",
                output_file="",
                notes=str(e)
            )))
            print(f"    -> exception    FAIL  ({e})")

    # Save & (optionally) display report
    if rows:
        report_path = Path(report_path)
        with report_path.open("w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
            writer.writeheader()
            for r in rows:
                writer.writerow(r)
        print(f"\nAggregated report written to: {report_path}")
        if PANDAS_OK:
            df = pd.DataFrame(rows)
            try:
                display_dataframe_to_user("MP4 Repair Report (from list)", df)
            except Exception:
                display(df)
        return rows
    else:
        print("No results to report.")
        return None


# ---------- Run if `videostofix` already exists in your notebook ----------
try:
    _ = notebook_repair_list(videostofix)
except NameError:
    print("Define `videostofix = [ ... ]` (list of full file paths) and re-run this cell.")
```

```python
# Deep diagnosis + more aggressive salvage attempts for specific files in `videostofix`.
# Run this cell AFTER defining `videostofix = [ ... ]` in your notebook.
#
# What this adds vs the previous cell:
# - Prints a diagnostic dataframe: size, presence of MP4 atoms (ftyp/moov/mdat), ffprobe + OpenCV status
# - Tries a more aggressive re-encode with larger probe/analyze windows and error-ignoring flags
# - Tries keyframe-only extraction/rebuild as a last-ditch salvage
#
# Outputs:
# - Per-file fixed outputs go into a `fixed/` subfolder next to each source
# - CSV: `deep_repair_report.csv` in the current working directory
#
# Requirements: ffmpeg + ffprobe on PATH. OpenCV optional.

import os, io, csv, json, shutil, tempfile, subprocess, struct
from pathlib import Path
from typing import Optional, Tuple, List, Dict
from dataclasses import dataclass, asdict

try:
    import pandas as pd
    from caas_jupyter_tools import display_dataframe_to_user
    PANDAS_OK = True
except Exception:
    PANDAS_OK = False

FPS_FALLBACK = 30
REPORT_PATH = Path("deep_repair_report.csv")

def which(cmd: str) -> Optional[str]:
    from shutil import which as _which
    return _which(cmd)

def run(cmd: List[str], timeout: Optional[int] = None) -> Tuple[int, str, str]:
    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    try:
        out, err = p.communicate(timeout=timeout)
    except subprocess.TimeoutExpired:
        p.kill()
        out, err = p.communicate()
    return p.returncode, out, err

def ffprobe_json(input_path: Path) -> Tuple[Optional[dict], str]:
    if not which("ffprobe"):
        return None, "ffprobe not found on PATH"
    cmd = [
        "ffprobe", "-v", "error", "-print_format", "json",
        "-show_format", "-show_streams", str(input_path)
    ]
    code, out, err = run(cmd)
    if code != 0:
        return None, (err.strip() or f"ffprobe exit code {code}")
    try:
        info = json.loads(out)
        return info, ""
    except Exception as e:
        return None, f"ffprobe json parse error: {e}"

def cv2_validate(input_path: Path, test_frames: int = 15) -> Tuple[bool, str, int, float]:
    try:
        import cv2
    except Exception:
        return False, "OpenCV not installed", 0, 0.0
    cap = cv2.VideoCapture(str(input_path))
    if not cap.isOpened():
        return False, "cv2.VideoCapture failed to open", 0, 0.0
    fps = cap.get(cv2.CAP_PROP_FPS) or 0.0
    frames_ok = 0
    reason = "ok"
    for i in range(test_frames):
        ok, frame = cap.read()
        if not ok or frame is None:
            reason = f"read stopped at frame {i}"
            break
        frames_ok += 1
    cap.release()
    return frames_ok > 0, reason, frames_ok, float(fps)

def scan_atoms(path: Path, max_bytes: int = 2_000_000) -> Dict[str, bool]:
    """Quick-and-dirty scan for MP4 atoms 'ftyp','moov','mdat' within the first `max_bytes` (and last chunk for 'moov')."""
    result = {"ftyp": False, "moov": False, "mdat": False}
    try:
        sz = path.stat().st_size
        with open(path, "rb") as f:
            head = f.read(min(max_bytes, sz))
            if b"ftyp" in head: result["ftyp"] = True
            if b"mdat" in head: result["mdat"] = True
            if b"moov" in head: result["moov"] = True
            if not result["moov"] and sz > max_bytes:
                # Also peek at the tail (moov can be at end)
                f.seek(max(0, sz - max_bytes))
                tail = f.read(max_bytes)
                if b"moov" in tail: result["moov"] = True
                if b"mdat" in tail and not result["mdat"]: result["mdat"] = True
    except Exception:
        pass
    return result

def try_reencode_aggressive(input_path: Path, out_path: Path, fps_fallback: int = 30) -> Tuple[bool, str]:
    """Aggressive ffmpeg re-encode with big probe/analyze and error-ignoring flags."""
    if not which("ffmpeg"): return False, "ffmpeg not found on PATH"
    cmd = [
        "ffmpeg",
        "-v", "error",
        "-probesize", "500M",
        "-analyzeduration", "500M",
        "-fflags", "+igndts+genpts+discardcorrupt",
        "-err_detect", "ignore_err",
        "-use_wallclock_as_timestamps", "1",
        "-i", str(input_path),
        "-map", "0:v:0",     # video only
        "-an",
        "-c:v", "libx264",
        "-pix_fmt", "yuv420p",
        "-vsync", "2",
        "-r", str(fps_fallback),
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, "aggressive reencode ok"
    return False, err.strip() or "aggressive reencode failed"

def try_keyframe_extract_rebuild(input_path: Path, out_path: Path, fps_fallback: int = 30, tmp_dir: Optional[Path]=None) -> Tuple[bool, str]:
    """Extract keyframes only (more tolerant) then rebuild a playable video."""
    if not which("ffmpeg"): return False, "ffmpeg not found on PATH"
    tmp = tmp_dir or Path(tempfile.mkdtemp(prefix="mp4kf_"))
    frames_dir = tmp / "kf"
    frames_dir.mkdir(parents=True, exist_ok=True)
    # Extract I-frames only; ignore errors
    cmd_extract = [
        "ffmpeg", "-v", "error",
        "-probesize", "500M",
        "-analyzeduration", "500M",
        "-fflags", "+igndts+genpts+discardcorrupt",
        "-err_detect", "ignore_err",
        "-i", str(input_path),
        "-vf", "select='eq(pict_type\\,I)'",
        "-vsync", "0",
        str(frames_dir / "%06d.png")
    ]
    code, out, err = run(cmd_extract)
    extracted = sorted(frames_dir.glob("*.png"))
    if len(extracted) == 0:
        shutil.rmtree(tmp, ignore_errors=True)
        return False, err.strip() or "no keyframes extracted"
    # Rebuild
    cmd_build = [
        "ffmpeg", "-v", "error",
        "-framerate", str(fps_fallback),
        "-i", str(frames_dir / "%06d.png"),
        "-c:v", "libx264",
        "-pix_fmt", "yuv420p",
        "-movflags", "+faststart",
        str(out_path)
    ]
    code, out, err = run(cmd_build)
    shutil.rmtree(tmp, ignore_errors=True)
    if code == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, f"rebuilt from {len(extracted)} keyframes"
    return False, err.strip() or "keyframe rebuild failed"

@dataclass
class DeepReport:
    source: str
    exists: bool
    size_bytes: int
    has_ftyp: bool
    has_moov: bool
    has_mdat: bool
    ffprobe_ok: bool
    ffprobe_err: str
    opencv_ok: bool
    opencv_reason: str
    action: str
    ok_fixed: bool
    reason_fixed: str
    output_file: str
    notes: str

def deep_repair_list(file_list: List[str],
                     fps_fallback: int = FPS_FALLBACK,
                     report_path: Path = REPORT_PATH):
    rows = []
    paths = [Path(p) for p in file_list]
    print(f"Deep-diagnosing {len(paths)} file(s)â€¦\n")
    for idx, p in enumerate(paths, 1):
        exists = p.exists()
        size = p.stat().st_size if exists else 0
        atoms = scan_atoms(p) if exists else {"ftyp": False, "moov": False, "mdat": False}
        info, probe_err = (None, "missing") if not exists else ffprobe_json(p)
        ff_ok = info is not None and probe_err == ""
        cv_ok, cv_reason, _, _ = (False, "missing", 0, 0.0) if not exists else cv2_validate(p)

        action = "none"
        ok_fixed = False
        reason_fixed = ""
        out_file = ""

        print(f"[{idx}] {p.name}  size={size:,}  atoms={atoms}  ffprobe_ok={ff_ok}  cv={cv_ok}")
        if not exists or size < 2048:
            reason_fixed = "file missing or effectively empty (<2 KB) â€” no salvage possible"
        else:
            out_dir = p.parent / "fixed"
            out_dir.mkdir(parents=True, exist_ok=True)
            # If no moov but mdat present, try aggressive decode-based salvage
            if atoms.get("mdat", False):
                # 1) Aggressive re-encode
                dst1 = out_dir / (p.stem + ".aggr.mp4")
                ok1, msg1 = try_reencode_aggressive(p, dst1, fps_fallback=fps_fallback)
                if ok1:
                    action, ok_fixed, reason_fixed, out_file = "aggressive-reencode", True, msg1, str(dst1)
                else:
                    # 2) Keyframe-only salvage
                    dst2 = out_dir / (p.stem + ".keyframes.mp4")
                    ok2, msg2 = try_keyframe_extract_rebuild(p, dst2, fps_fallback=fps_fallback)
                    if ok2:
                        action, ok_fixed, reason_fixed, out_file = "keyframe-rebuild", True, msg2, str(dst2)
                    else:
                        action, ok_fixed, reason_fixed = "failed-all", False, f"{msg1} | {msg2}"
            else:
                reason_fixed = "no 'mdat' payload detected in quick scan â€” likely zeroed/truncated"

        rows.append(asdict(DeepReport(
            source=str(p),
            exists=exists,
            size_bytes=size,
            has_ftyp=atoms.get("ftyp", False),
            has_moov=atoms.get("moov", False),
            has_mdat=atoms.get("mdat", False),
            ffprobe_ok=ff_ok,
            ffprobe_err="" if ff_ok else probe_err,
            opencv_ok=cv_ok,
            opencv_reason=cv_reason,
            action=action,
            ok_fixed=ok_fixed,
            reason_fixed=reason_fixed,
            output_file=out_file,
            notes=("Consider 'untrunc' with a good reference file from the same camera if mdat present but moov missing."
                   if (exists and atoms.get('mdat', False) and not atoms.get('moov', False) and not ok_fixed) else "")
        )))

    # Save and display
    if rows:
        with open(report_path, "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=list(rows[0].keys()))
            writer.writeheader()
            for r in rows:
                writer.writerow(r)
        print(f"\nDeep report written to: {report_path}")
        if PANDAS_OK:
            df = pd.DataFrame(rows)
            try:
                display_dataframe_to_user("Deep MP4 Diagnosis & Repair", df)
            except Exception:
                display(df)
    else:
        print("No rows to report.")
    return rows

# ---- Run if videostofix exists ----
try:
    deep_rows = deep_repair_list(videostofix, fps_fallback=FPS_FALLBACK, report_path=REPORT_PATH)
except NameError:
    print("Define `videostofix = [ ... ]` and re-run this cell.")
```

```python
# Jupyter cell: Rescue MP4s missing the 'moov' atom using a GOOD reference file (via `untrunc`),
# with automatic validation and a fallback (keyframe-only rebuild) if `untrunc` isn't available.
#
# HOW TO USE:
# 1) Define your list of broken files as `videostofix = [ ... ]` (full paths)
# 2) Set REF_GOOD to a *healthy* MP4 from the same setup (e.g., one of your files 8 or 9)
# 3) Optionally set UNTRUNC_PATH to where you installed untrunc (on Windows: r"C:\tools\untrunc\untrunc.exe").
# 4) Run this cell.
#
# Outputs:
# - Fixed files go into a `fixed` subfolder next to EACH broken file
# - A CSV report `untrunc_repair_report.csv` is written in the current working directory
#
# Requirements:
# - `untrunc` installed (recommended) -> https://github.com/ponchio/untrunc (or prebuilt binary)
# - `ffmpeg`/`ffprobe` installed (for validation and fallback)
# - OpenCV optional (pip install opencv-python-headless) for extra validation

import csv
import json
import time
import shutil
import subprocess
from pathlib import Path
from dataclasses import dataclass, asdict
from typing import Optional, Tuple, List, Dict

# ---- USER SETTINGS (EDIT THESE) ----
try:
    videostofix  # noqa: F821  # must be defined by you in another cell
except NameError:
    videostofix = []  # placeholder; define it in another cell

REF_GOOD = r"C:/Users/aleja/Box/Awake Project/Maze data/Auditory experiments/8_arms_w_voc (1)/w1_d1/time_2025-06-04_14_22_30mouse10049/mouse10049_time_2025-06-04_14_22_30.mp4" 

# If untrunc is on PATH, leave as None. Otherwise, set the full path to the binary (Windows example below).
UNTRUNC_PATH = "C:/Users/aleja/Downloads/untrunc_x64/untrunc_x64/untrunc.exe"  # e.g., r"C:\tools\untrunc\untrunc.exe" or "/usr/local/bin/untrunc"

# Fallback behavior if untrunc is not available
ENABLE_KEYFRAME_FALLBACK = True
FPS_FALLBACK = 30

REPORT_CSV = Path("untrunc_repair_report.csv")
# ------------------------------------

# Optional: nice DataFrame display if pandas is available
try:
    import pandas as pd
    from caas_jupyter_tools import display_dataframe_to_user
    PANDAS_OK = True
except Exception:
    PANDAS_OK = False


def which(cmd: str) -> Optional[str]:
    from shutil import which as _which
    return _which(cmd)


def run(cmd: List[str], cwd: Optional[Path] = None, timeout: Optional[int] = None) -> Tuple[int, str, str]:
    p = subprocess.Popen(cmd, cwd=str(cwd) if cwd else None,
                         stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    try:
        out, err = p.communicate(timeout=timeout)
    except subprocess.TimeoutExpired:
        p.kill()
        out, err = p.communicate()
    return p.returncode, out, err


def ffprobe_json(input_path: Path) -> Tuple[Optional[dict], str]:
    if which("ffprobe") is None:
        return None, "ffprobe not found on PATH"
    code, out, err = run([
        "ffprobe", "-v", "error", "-print_format", "json",
        "-show_format", "-show_streams", str(input_path)
    ])
    if code != 0:
        return None, (err.strip() or f"ffprobe exit code {code}")
    try:
        info = json.loads(out)
        return info, ""
    except Exception as e:
        return None, f"ffprobe json parse error: {e}"


def cv2_validate(input_path: Path, test_frames: int = 15) -> Tuple[bool, str, int, float]:
    try:
        import cv2
    except Exception:
        return False, "OpenCV not installed", 0, 0.0
    cap = cv2.VideoCapture(str(input_path))
    if not cap.isOpened():
        return False, "cv2.VideoCapture failed to open", 0, 0.0
    fps = cap.get(cv2.CAP_PROP_FPS) or 0.0
    frames_ok = 0
    reason = "ok"
    for i in range(test_frames):
        ok, frame = cap.read()
        if not ok or frame is None:
            reason = f"read stopped at frame {i}"
            break
        frames_ok += 1
    cap.release()
    return frames_ok > 0, reason, frames_ok, float(fps)


def scan_atoms(path: Path, max_bytes: int = 2_000_000) -> Dict[str, bool]:
    result = {"ftyp": False, "moov": False, "mdat": False}
    try:
        sz = path.stat().st_size
        with open(path, "rb") as f:
            head = f.read(min(max_bytes, sz))
            if b"ftyp" in head: result["ftyp"] = True
            if b"mdat" in head: result["mdat"] = True
            if b"moov" in head: result["moov"] = True
            if not result["moov"] and sz > max_bytes:
                f.seek(max(0, sz - max_bytes))
                tail = f.read(max_bytes)
                if b"moov" in tail: result["moov"] = True
                if b"mdat" in tail and not result["mdat"]: result["mdat"] = True
    except Exception:
        pass
    return result


def try_keyframe_extract_rebuild(input_path: Path, out_path: Path, fps_fallback: int = 30) -> Tuple[bool, str]:
    """Last-ditch: extract I-frames and rebuild; not perfect but may salvage visuals."""
    if which("ffmpeg") is None:
        return False, "ffmpeg not found on PATH"
    # 1) Extract keyframes to PNG
    tmp_dir = out_path.parent / f"tmp_kf_{int(time.time())}"
    frames_dir = tmp_dir / "kf"
    frames_dir.mkdir(parents=True, exist_ok=True)
    code, out, err = run([
        "ffmpeg", "-v", "error",
        "-probesize", "500M", "-analyzeduration", "500M",
        "-fflags", "+igndts+genpts+discardcorrupt",
        "-err_detect", "ignore_err",
        "-i", str(input_path),
        "-vf", "select=eq(pict_type\\,I)",
        "-vsync", "0",
        str(frames_dir / "%06d.png")
    ])
    extracted = sorted(frames_dir.glob("*.png"))
    if len(extracted) == 0:
        shutil.rmtree(tmp_dir, ignore_errors=True)
        return False, (err.strip() or "no keyframes extracted")

    # 2) Rebuild video
    code2, out2, err2 = run([
        "ffmpeg", "-v", "error",
        "-framerate", str(fps_fallback),
        "-i", str(frames_dir / "%06d.png"),
        "-c:v", "libx264",
        "-pix_fmt", "yuv420p",
        "-movflags", "+faststart",
        str(out_path)
    ])
    shutil.rmtree(tmp_dir, ignore_errors=True)
    if code2 == 0 and out_path.exists() and out_path.stat().st_size > 0:
        return True, f"rebuilt from {len(extracted)} keyframes"
    return False, (err2.strip() or "keyframe rebuild failed")


def run_untrunc(reference_good: Path, broken: Path, out_dir: Path, untrunc_path: Optional[str]) -> Tuple[bool, str, Optional[Path]]:
    """Run untrunc in out_dir, detect the output file, and return its path."""
    exe = untrunc_path or which("untrunc") or which("untrunc.exe")
    if exe is None:
        return False, "untrunc not found on PATH and UNTRUNC_PATH not set", None

    out_dir.mkdir(parents=True, exist_ok=True)
    before = {p.name for p in out_dir.glob("*.mp4")}
    start = time.time()

    code, out, err = run([exe, str(reference_good), str(broken)], cwd=out_dir)
    # Detect new mp4 created after running untrunc
    candidates = [p for p in out_dir.glob("*.mp4") if p.name not in before and p.stat().st_mtime >= start]
    if code != 0 and not candidates:
        return False, (err.strip() or "untrunc failed, no output produced"), None
    if not candidates:
        # Some builds may overwrite or reuse names; consider any file touched after start
        candidates = [p for p in out_dir.glob("*.mp4") if p.stat().st_mtime >= start]

    if not candidates:
        return False, "untrunc produced no detectable output", None

    # Choose the largest newest candidate
    best = sorted(candidates, key=lambda p: (p.stat().st_mtime, p.stat().st_size), reverse=True)[0]
    return True, "untrunc produced output", best


@dataclass
class Row:
    source: str
    size_bytes: int
    has_ftyp: bool
    has_moov: bool
    has_mdat: bool
    action: str
    ok: bool
    msg: str
    output_file: str
    probe_ok: bool
    probe_err: str
    opencv_ok: bool
    opencv_note: str


def rescue_with_untrunc(videopaths: List[str], ref_good: str, untrunc_path: Optional[str] = None) -> List[Row]:
    ref = Path(ref_good)
    if not ref.exists():
        raise FileNotFoundError(f"Reference GOOD file not found: {ref}")
    rows: List[Row] = []

    for idx, s in enumerate(videopaths, 1):
        src = Path(s)
        print(f"[{idx}/{len(videopaths)}] {src.name}")
        if not src.exists():
            print("   -> missing, skipped")
            rows.append(Row(str(src), 0, False, False, False, "skip-missing", False, "file not found", "", False, "missing", False, "missing"))
            continue

        atoms = scan_atoms(src)
        size = src.stat().st_size

        out_dir = src.parent / "fixed"
        out_dir.mkdir(parents=True, exist_ok=True)

        # Only use untrunc if moov is missing but mdat is present
        if atoms.get("mdat", False) and not atoms.get("moov", False):
            ok, msg, out_path = run_untrunc(ref, src, out_dir, untrunc_path)
            if ok and out_path and out_path.exists():
                # Optional: normalize / faststart remux
                normalized = out_dir / (src.stem + ".untrunc.mp4")
                # If ffmpeg exists, remux to a stable, known name
                if which("ffmpeg"):
                    code, out, err = run([
                        "ffmpeg", "-v", "error",
                        "-i", str(out_path),
                        "-c", "copy",
                        "-movflags", "+faststart",
                        str(normalized)
                    ])
                    if code == 0 and normalized.exists() and normalized.stat().st_size > 0:
                        out_path = normalized

                # Validate
                info, perr = ffprobe_json(out_path)
                probe_ok = info is not None and perr == ""
                cv_ok, cv_note, _, _ = cv2_validate(out_path)

                print(f"   -> untrunc OK  -> {out_path.name}")
                rows.append(Row(str(src), size, atoms["ftyp"], atoms["moov"], atoms["mdat"],
                                "untrunc", True, msg, str(out_path),
                                probe_ok, "" if probe_ok else perr, cv_ok, cv_note))
            else:
                print(f"   -> untrunc FAILED: {msg}")
                if ENABLE_KEYFRAME_FALLBACK:
                    kf_out = out_dir / (src.stem + ".keyframes.mp4")
                    ok2, msg2 = try_keyframe_extract_rebuild(src, kf_out, fps_fallback=FPS_FALLBACK)
                    info, perr = ffprobe_json(kf_out) if ok2 else (None, "no output")
                    cv_ok, cv_note, _, _ = cv2_validate(kf_out) if ok2 else (False, "no output", 0, 0.0)
                    rows.append(Row(str(src), size, atoms["ftyp"], atoms["moov"], atoms["mdat"],
                                    "keyframe-rebuild", ok2, msg2, str(kf_out if ok2 else ""),
                                    info is not None and perr == "", "" if info else perr, cv_ok, cv_note))
                else:
                    rows.append(Row(str(src), size, atoms["ftyp"], atoms["moov"], atoms["mdat"],
                                    "untrunc", False, msg, "",
                                    False, "not validated", False, "not validated"))
        else:
            # If moov present (or mdat missing), we won't untruncâ€”just report
            info, perr = ffprobe_json(src)
            probe_ok = info is not None and perr == ""
            cv_ok, cv_note, _, _ = cv2_validate(src)
            rows.append(Row(str(src), size, atoms["ftyp"], atoms["moov"], atoms["mdat"],
                            "no-action", True, "already has moov or no mdat", str(src),
                            probe_ok, "" if probe_ok else perr, cv_ok, cv_note))

    # Write report
    if rows:
        with REPORT_CSV.open("w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=list(asdict(rows[0]).keys()))
            writer.writeheader()
            for r in rows:
                writer.writerow(asdict(r))
        print(f"\nReport written to: {REPORT_CSV}")
        if PANDAS_OK:
            df = pd.DataFrame([asdict(r) for r in rows])
            try:
                display_dataframe_to_user("Untrunc Repair Report", df)
            except Exception:
                display(df)
    return rows


# ---- RUN (will only run if you have defined `videostofix` and set REF_GOOD) ----
try:
    if not videostofix:
        print("Define `videostofix = [ ... ]` with your broken file paths and set REF_GOOD, then re-run this cell.")
    elif not Path(REF_GOOD).exists():
        print(f"Set REF_GOOD to a valid healthy MP4 (e.g., your items 8 or 9). Current REF_GOOD not found: {REF_GOOD}")
    else:
        _rows = rescue_with_untrunc(videostofix, REF_GOOD, UNTRUNC_PATH)
except Exception as e:
    print("Error while running untrunc rescue:", e)
```

```python
# Jupyter cell: Next-level salvage for MP4s missing the 'moov' atom
# - Carves 'mdat', sniffs codec signatures, tries raw decodes (m4v/h264/hevc/mjpeg)
# - Also tries a heuristic conversion of length-prefixed H.264 to Annex B before decoding
# - Validates outputs and writes a CSV + shows an interactive table (if pandas is available)
#
# Use when: untrunc failed and keyframe rebuild couldn't even open the file.
#
# HOW TO RUN:
# 1) Ensure `videostofix = [ ... ]` lists your broken files (the 7 with moov âŒ).
# 2) Run this cell. Outputs go to a `fixed/` folder next to each source. Report: `mdat_salvage_v2_report.csv`.
#
import os, io, csv, json, struct, shutil, tempfile, subprocess
from pathlib import Path
from dataclasses import dataclass, asdict
from typing import Optional, Tuple, List, Dict

REPORT_CSV = Path("mdat_salvage_v2_report.csv")
FPS_FALLBACK = 30
KEEP_RAW = False  # set True to inspect the carved payloads

# Optional: DataFrame display
try:
    import pandas as pd
    from caas_jupyter_tools import display_dataframe_to_user
    PANDAS_OK = True
except Exception:
    PANDAS_OK = False

def which(cmd: str) -> Optional[str]:
    from shutil import which as _which
    return _which(cmd)

def run(cmd: List[str]) -> Tuple[int, str, str]:
    p = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    out, err = p.communicate()
    return p.returncode, out, err

def ffprobe_json(path: Path) -> Tuple[Optional[dict], str]:
    if which("ffprobe") is None:
        return None, "ffprobe not found on PATH"
    code, out, err = run(["ffprobe","-v","error","-print_format","json","-show_format","-show_streams",str(path)])
    if code != 0:
        return None, (err.strip() or f"ffprobe exit {code}")
    try:
        return json.loads(out), ""
    except Exception as e:
        return None, f"ffprobe json parse error: {e}"

def cv2_validate(path: Path, n: int = 10) -> Tuple[bool, str]:
    try:
        import cv2
    except Exception:
        return False, "OpenCV not installed"
    cap = cv2.VideoCapture(str(path))
    if not cap.isOpened():
        return False, "cv2.VideoCapture failed"
    ok = 0
    for i in range(n):
        r, f = cap.read()
        if not r or f is None: break
        ok += 1
    cap.release()
    return (ok > 0), (f"read {ok} frames" if ok > 0 else "no frames read")

# --------- MP4 parsing to find MDAT ---------
def parse_boxes_for_mdat(path: Path) -> List[Tuple[int, int]]:
    """
    Return list of (data_offset, data_size) for top-level 'mdat' boxes.
    Handles 32-bit size, 64-bit extended size, and size==0 (to EOF).
    """
    mdats = []
    with open(path, "rb") as f:
        filesize = f.seek(0, os.SEEK_END)
        f.seek(0)
        pos = 0
        while pos + 8 <= filesize:
            f.seek(pos)
            header = f.read(8)
            if len(header) < 8:
                break
            size, typ = struct.unpack(">I4s", header)
            typ = typ.decode("latin1")
            header_size = 8
            if size == 1:
                if pos + 16 > filesize: break
                largesize = struct.unpack(">Q", f.read(8))[0]
                header_size = 16
                box_size = largesize
            elif size == 0:
                box_size = filesize - pos
            else:
                box_size = size

            if box_size < header_size:
                break

            if typ == "mdat":
                data_offset = pos + header_size
                data_size = box_size - header_size
                if data_offset + data_size > filesize:
                    data_size = max(0, filesize - data_offset)
                mdats.append((data_offset, data_size))

            if box_size == 0:
                break
            pos += box_size
    return mdats

def carve_mdat_to_temp(path: Path, out_dir: Path) -> Optional[Path]:
    mdats = parse_boxes_for_mdat(path)
    if not mdats:
        return None
    out_dir.mkdir(parents=True, exist_ok=True)
    raw_path = out_dir / (path.stem + ".mdat.bin")
    with open(path, "rb") as fin, open(raw_path, "wb") as fout:
        for (off, size) in mdats:
            fin.seek(off)
            to_copy = size
            chunk = 4 * 1024 * 1024
            while to_copy > 0:
                data = fin.read(min(chunk, to_copy))
                if not data: break
                fout.write(data)
                to_copy -= len(data)
    return raw_path if raw_path.exists() and raw_path.stat().st_size > 0 else None

# --------- Signature sniffing ---------
def sniff_payload_signatures(raw_path: Path, sample_bytes: int = 4_000_000) -> Dict[str, int]:
    """
    Count occurrences of common start codes:
      - MPEG-4 Visual: 0x000001B0 (VO), 0x000001B6 (VOP)  -> keys: 'm4v_b0', 'm4v_b6'
      - Annex B H.264:  00 00 01 [65,67,68]               -> keys: 'h264_sps','h264_pps','h264_idr'
    This is heuristic but useful to guide attempts.
    """
    sigs = {"m4v_b0":0, "m4v_b6":0, "h264_sps":0, "h264_pps":0, "h264_idr":0}
    try:
        with open(raw_path, "rb") as f:
            data = f.read(sample_bytes)
        # MPEG-4 Visual
        sigs["m4v_b0"] = data.count(b"\x00\x00\x01\xb0")
        sigs["m4v_b6"] = data.count(b"\x00\x00\x01\xb6")
        # H.264 Annex B (00 00 01)
        # NAL types: 7=SPS, 8=PPS, 5=IDR
        sigs["h264_sps"] = data.count(b"\x00\x00\x01\x67")
        sigs["h264_pps"] = data.count(b"\x00\x00\x01\x68")
        sigs["h264_idr"] = data.count(b"\x00\x00\x01\x65")
    except Exception:
        pass
    return sigs

# --------- Heuristic conversion: length-prefixed H.264 -> Annex B ---------
def try_lenpref_h264_to_annexb(raw_path: Path, out_annexb: Path, max_frames: int = 10_000) -> Tuple[bool, str]:
    """
    Assume 4-byte big-endian length prefixes for NAL units.
    We'll parse sequentially and write Annex B start codes. If it looks implausible, we abort.
    Returns (ok, message).
    """
    try:
        with open(raw_path, "rb") as fin, open(out_annexb, "wb") as fout:
            size = fin.seek(0, os.SEEK_END)
            fin.seek(0)
            pos = 0
            frames = 0
            while pos + 4 <= size and frames < max_frames:
                len_bytes = fin.read(4)
                if len(len_bytes) < 4:
                    break
                nalu_len = int.from_bytes(len_bytes, "big", signed=False)
                pos += 4
                if nalu_len <= 0 or nalu_len > (size - pos) or nalu_len > 10_000_000:
                    # Implausible -> probably not length-prefixed H.264
                    return False, f"implausible NAL length at pos {pos-4}: {nalu_len}"
                nalu = fin.read(nalu_len)
                if len(nalu) != nalu_len:
                    return False, "short read for NAL"
                # write Annex B start code + NAL
                fout.write(b"\x00\x00\x00\x01")
                fout.write(nalu)
                pos += nalu_len
                frames += 1
        return True, f"converted {frames} NAL units to Annex B"
    except Exception as e:
        return False, f"exception: {e}"

# --------- Decode attempts ---------
def try_decode_raw(raw_path: Path, out_mp4: Path, fmt: str, fps: int = 30) -> Tuple[bool, str]:
    if which("ffmpeg") is None:
        return False, "ffmpeg not found on PATH"
    cmd = [
        "ffmpeg","-v","error",
        "-probesize","500M","-analyzeduration","500M",
        "-fflags","+igndts+genpts+discardcorrupt",
        "-err_detect","ignore_err",
        "-f", fmt, "-i", str(raw_path),
        "-an",
        "-c:v","libx264","-pix_fmt","yuv420p",
        "-vsync","2","-r", str(fps),
        "-movflags","+faststart",
        str(out_mp4)
    ]
    code, out, err = run(cmd)
    if code == 0 and out_mp4.exists() and out_mp4.stat().st_size > 0:
        return True, f"decoded as {fmt} -> {out_mp4.name}"
    return False, (err.strip() or f"decode failed as {fmt}")

@dataclass
class Row:
    source: str
    carved_bytes: int
    sig_m4v_b0: int
    sig_m4v_b6: int
    sig_h264_sps: int
    sig_h264_pps: int
    sig_h264_idr: int
    strategy: str
    ok: bool
    msg: str
    output_file: str
    probe_ok: bool
    probe_err: str
    opencv_ok: bool
    opencv_note: str

def salvage_from_list(paths: List[str], fps_fallback: int = FPS_FALLBACK, keep_raw: bool = KEEP_RAW):
    rows: List[Row] = []
    for idx, s in enumerate(paths, 1):
        src = Path(s)
        print(f"[{idx}/{len(paths)}] {src.name}")
        if not src.exists():
            print("   -> missing")
            rows.append(Row(str(src), 0, 0,0,0,0,0, "skip", False, "missing", "", False, "missing", False, "missing"))
            continue

        out_dir = src.parent / "fixed"
        out_dir.mkdir(parents=True, exist_ok=True)
        tmp_dir = out_dir / "_tmp"
        tmp_dir.mkdir(parents=True, exist_ok=True)

        raw = carve_mdat_to_temp(src, tmp_dir)
        if not raw:
            print("   -> no mdat carved")
            rows.append(Row(str(src), 0, 0,0,0,0,0, "carve", False, "no mdat carved", "", False, "n/a", False, "n/a"))
            continue

        raw_sz = raw.stat().st_size
        sigs = sniff_payload_signatures(raw)
        print(f"   -> carved {raw_sz/1e6:.1f} MB | sigs: {sigs}")

        # Decide attempt order
        attempts: List[Tuple[str, Path]] = []
        if (sigs["m4v_b0"] + sigs["m4v_b6"]) > 0:
            attempts.append(("m4v", out_dir / (src.stem + ".from_m4v.mp4")))
        # We'll try raw annexb next if we saw SPS/PPS/IDR
        if (sigs["h264_sps"] + sigs["h264_pps"] + sigs["h264_idr"]) > 0:
            attempts.append(("h264", out_dir / (src.stem + ".from_h264.mp4")))
        # Always include a couple more generic attempts
        attempts += [("hevc", out_dir / (src.stem + ".from_hevc.mp4")),
                     ("mjpeg", out_dir / (src.stem + ".from_mjpeg.mp4"))]

        success = False
        last_msg = ""
        out_file = ""

        # Try direct decodes
        for fmt, dst in attempts:
            ok, msg = try_decode_raw(raw, dst, fmt=fmt, fps=fps_fallback)
            print(f"      try {fmt:<5} -> {'OK' if ok else 'FAIL'}")
            last_msg = msg
            if ok:
                success, out_file = True, str(dst)
                break

        # If direct attempts failed, try heuristic H.264 length->AnnexB then decode
        if not success:
            ann = tmp_dir / (src.stem + ".annexb.h264")
            ok_c, msg_c = try_lenpref_h264_to_annexb(raw, ann)
            if ok_c and ann.exists():
                dst = out_dir / (src.stem + ".from_lenpref_annexb.mp4")
                ok, msg = try_decode_raw(ann, dst, fmt="h264", fps=fps_fallback)
                print(f"      lenprefâ†’annexb -> {'OK' if ok else 'FAIL'}")
                last_msg = msg if not ok else "converted length-prefixed H.264 to AnnexB"
                if ok:
                    success, out_file = True, str(dst)

        # Validate and finalize row
        probe_ok = False; probe_err = ""; cv_ok = False; cv_note = ""
        if success:
            info, perr = ffprobe_json(Path(out_file))
            probe_ok = info is not None and perr == ""
            probe_err = "" if probe_ok else perr
            cv_ok, cv_note = cv2_validate(Path(out_file))
            print(f"   -> SUCCESS: {Path(out_file).name}")
        else:
            print(f"   -> FAILED: {last_msg}")

        if not keep_raw:
            shutil.rmtree(tmp_dir, ignore_errors=True)

        rows.append(Row(str(src), raw_sz, sigs["m4v_b0"], sigs["m4v_b6"],
                        sigs["h264_sps"], sigs["h264_pps"], sigs["h264_idr"],
                        "mdat-salvage-v2", success, last_msg if last_msg else ("ok" if success else "failed"),
                        out_file if success else "",
                        probe_ok, probe_err, cv_ok, cv_note))

    # Report
    if rows:
        with REPORT_CSV.open("w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=list(asdict(rows[0]).keys()))
            writer.writeheader()
            for r in rows:
                writer.writerow(asdict(r))
        print(f"\nReport written to: {REPORT_CSV}")
        if PANDAS_OK:
            df = pd.DataFrame([asdict(r) for r in rows])
            try:
                display_dataframe_to_user("MDAT Salvage v2 Report", df)
            except Exception:
                display(df)
    return rows

    # ---- RUN on your existing `videost
try:
    _rows = salvage_from_list(videostofix, fps_fallback=FPS_FALLBACK, keep_raw=KEEP_RAW)
except NameError:
    print("Define `videostofix = [ ... ]` with your broken MP4 paths and re-run this cell.")
```
