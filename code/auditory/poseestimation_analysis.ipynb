```python
import deeplabcut as dlc
from pathlib import Path
import pandas as pd
import ipywidgets as widgets
```

```python
# create videos path
#for now we are just going to look at mouse 6357, session 3.6 since we have that fully labeled



mouse6357_session3_6_video_paths= []

path = "C:/Users/shahd/Box/Awake Project/Maze data/simplermaze/mouse 6357/2024-08-28_11_58_146357session3.6/mouse6357_session3.6_trial_info.csv"
mouse = pd.read_csv(path)

# for i in mouse["video_segment_path"]:
#     if type(i) != float:
#         mouse6357_session3_6.append(i)
#         print(i)

# mouse[['state_sequence', 'video_segment_path']]

mouset = mouse.dropna(axis = 0, subset=['state_sequence'] )

mouset = mouset[['state_sequence', 'video_segment_path']]

for i in mouset['video_segment_path']:
    # print(i)
    mouse6357_session3_6_video_paths.append(i)

print(len(mouse6357_session3_6_video_paths))



```

```python


# Choose a folder to host the DLC project (outside your data)
project_name = "MiceMaze"
your_name = "Alej"
videos = mouse6357_session3_6_video_paths

config_path = dlc.create_new_project(
    project=project_name,
    experimenter=your_name,
    videos=videos,
    working_directory=r"C:/Users/shahd/OneDrive/Documents/GitHub/mice-maze/code/simplermaze",  # pick a fast local SSD
    copy_videos=True,                     # keep videos in where they are
)
print("Config:", config_path)
```
